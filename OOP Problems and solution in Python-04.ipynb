A vehicle assumes that the whole world is a 2-dimensional graph paper. It maintains its<br>
x and y coordinates (both are integers). The vehicle gets manufactured (constructed) at<br>
(0, 0) coordinate.<br>
Subtasks:<br>
1. Design a Vehicle2010 class which inherits movement methods from Vehicle<br>
and adds new methods called move UpperRight, UpperLeft, LowerRight,<br>
LowerLeft. Each of these diagonal move methods must re-use two inherited and<br>
appropriate move methods.<br>
2. Write an “equals” method which tests if significant class properties are the<br>
same (in this case x and y).<br>
Note: All moves are 1 step. That means a single call to any move method changes<br>
value of either x or y or both by 1.<br>

```python
class Vehicle:
    def __init__(self):
        self.x = 0
        self.y = 0
    def moveUp(self):
        self.y+=1
    def moveDown(self):
        self.y-=1
    def moveRight(self):
        self.x+=1
    def moveLeft(self):
        self.x-=1
    def __str__(self):
        return '('+str(self.x)+' , '+str(self.y)+')'
class Vehicle2010(Vehicle):
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def moveLowerLeft(self):
        super().moveLeft()
        super().moveDown()
    def equals(self,other):
        return self.__dict__ == other.__dict__
        
print('Part 1')
print('------')
car = Vehicle()
print(car)
car.moveUp()
print(car)
car.moveLeft()
print(car)
car.moveDown()
print(car)
car.moveRight()
print(car)
print('------')
print('Part 2')
print('------')
car1 = Vehicle2010()
print(car1)
car1.moveLowerLeft()
print(car1)
car2 = Vehicle2010()
car2.moveLeft()
print(car1.equals(car2))
car2.moveDown()
print(car1.equals(car2))
```
