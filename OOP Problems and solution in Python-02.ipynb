# Question 1

Suppose your little sibling wants your help to check his math homework. He is done with<br>
his homework but wants you to see if all his results are correct. Since the student with<br>
all correct results gets 3 stars. However, you want your brother to check this on his own.<br>
So, you design a calculator for him in python. You could have given your scientific<br>
calculator but you wanted to give him a basic calculator and also wanted to see if you<br>
can even design one.<br>
Subtasks:<br>
1. Create a class called Calculator.<br>
2. Your class shall have 1 constructor and 4 methods, namely add, subtract,<br>
multiply and divide.<br>
3. Now, create an object of your class. After creating an object, it should print “Let’s<br>
Calculate!”<br>
4. Then take 3 inputs from the user: first value, operator, second value<br>
5. Now based on the given operator, call the required method and print the result.<br>
Sample Input:<br>
1<br>
+<br>
2<br>
Sample Output:<br>
Let’s Calculate!<br>
Value 1: 1<br>
Operator: +<br>
Value 2: 2<br>
Result: 3<br>

```python
class Calculator:
    def __init__(self,num1,operator,num2):
        self.num1=num1
        self.operator=operator
        self.num2=num2
        print('Let’s Calculate!')
    def add(self):
        return num1+num2
    def substract(self):
        return num1-num2
    def multiply(self):
        return num1*num2
    def divide(self):
        return num1/num2
num1=int(input())
operator=input()
num2=int(input())

if operator=='+':
    Result=Calculator(num1,operator,num2)
    print('Value 1:',num1)
    print('Operator:',operator)
    print('Value 2:',num2)
    print('Result:',Result.add())
if operator=='-':
    Result=Calculator(num1,operator,num2)
    print('Value 1:',num1)
    print('Operator:',operator)
    print('Value 2:',num2)
    print('Result:',Result.substract())
if operator=='*':
    Result=Calculator(num1,operator,num2)
    print('Value 1:',num1)
    print('Operator:',operator)
    print('Value 2:',num2)
    print('Result:',Result.multiply())
if operator=='/':
    Result=Calculator(num1,operator,num2)
    print('Value 1:',num1)
    print('Operator:',operator)
    print('Value 2:',num2)
    print('Result:',Result.divide())
```

Design a “Vehicle” class. A vehicle assumes that the whole world is a 2-dimensional<br>

It must have methods to move up, down, left, right and a print_position() method for<br>
printing the current coordinate.<br>
Note: All moves are 1 step. That means a single call to any move method changes the<br>
value of either x or y or both by 1.<br>

**Write your class here**<br>
car = Vehicle()<br>
car.print_position()<br>
car.moveUp()<br>
car.print_position()<br>
car.moveLeft()<br>
car.print_position()<br>

OUTPUT<br>

(-1,1)<br>

```python
class Vehicle:
    def __init__(self):
        self.x=0
        self.y=0
    def moveUp(self):
        self.x=self.x
        self.y=self.y+1
    def moveDown(self):
        self.x=self.x
        self.y=self.y-1
    def moveLeft(self):
        self.x=self.x-1
        self.y=self.y
    def moveRight(self):
        self.x=self.x+1
        self.y=self.y
    def print_position(self):
        print('('+str(self.x)+','+str(self.y)+')')
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```
